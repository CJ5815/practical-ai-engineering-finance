# Week 3: Performance Summary

Same logic as `performance_summary.py`, using the reusable, tested functions from `ai_finance_course.analysis` — this notebook just wires them together and displays the result.

```python
import pandas as pd
from matplotlib import pyplot as plt

from ai_finance_course.analysis import (
    add_daily_returns,
    cumulative_growth,
    rolling_average,
    rolling_volatility,
)

ROLLING_WINDOW = 3  # short window since the sample data only has 5 rows

prices = pd.read_csv("../../data/sample/prices.csv")
prices = add_daily_returns(prices)
prices["rolling_avg_return"] = rolling_average(prices["return"], ROLLING_WINDOW)
prices["rolling_volatility"] = rolling_volatility(prices["return"], ROLLING_WINDOW)
prices["cumulative_growth"] = cumulative_growth(prices["return"])

prices
```

## Plot Cumulative Growth

In the `.py` script this calls `plt.show()`, which opens a window. Here, the chart renders inline below the cell.

```python
plt.plot(prices["date"], prices["cumulative_growth"])
plt.title("DEMO Cumulative Growth of $1")
plt.xlabel("Date")
plt.ylabel("Growth Factor")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
